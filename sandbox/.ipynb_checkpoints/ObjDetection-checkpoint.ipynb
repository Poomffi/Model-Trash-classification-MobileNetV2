```python
model = YOLO("yolov8n.pt")
```

```python
def getPredictedImages(src_img):
    
    if os.path.exists('detection'):
        shutil.rmtree('detection')
        
    results = model.predict(source=src_img, project="detection", name="obj", save_crop=True, show_labels=True, save_txt=True)
    
    src_folder = 'detection/obj/crops'
    files = glob.glob(os.path.join(src_folder, '*', '*'))
    
    
    if os.path.exists('results'):
        shutil.rmtree('results')
    
    count = 1
    for file in files:
        
        dest_path = os.path.join('results', 'predict' + str(count) + '.jpg')
        
        os.makedirs(os.path.dirname(dest_path), exist_ok=True)

        shutil.move(file, dest_path)
    
        count += 1
    
    
#     if os.path.exists('detection'):
#         shutil.rmtree('detection')
```

```python
def getCropWithCood(src_img):
    
    if os.path.exists('detection'):
        shutil.rmtree('detection')
        
    results = model.predict(src_img)
    
#     for result_img in results:
#         continue
#     print(result_img.boxes.xyxy)
   
    boxes = results[0].boxes.xyxy.tolist()
    print(boxes)
    
    img = cv2.imread(src_img)
    
    directory_name = "result"

    if not os.path.exists(directory_name):
        os.makedirs(directory_name)
    
#     all_imgs = []
    
    for i, box in enumerate(boxes):
        x1, y1, x2, y2 = box
        
        cv2.rectangle(img, (int(x1), int(y1)), (int(x2), int(y2)), (0, 255, 0), 2)
        
#         detect_object = img[int(y1):int(y2), int(x1):int(x2)]
        
#         all_imgs.append(detect_object)
        
#         cv2.imwrite('result/detect_obg_' + str(i) + '.jpg', detect_object)
    
#     return all_imgs
    
    return img
    
```

```python
classifier = load_model('modelMobile.h5')
```

```python
img_height = 224
img_width = 224
```

```python
image_path = "./result/"
lists = os.listdir(image_path)
print(lists)
```

```python
categories = ['battery', 'biological', 'brown-glass', 'cardboard', 'clothes', 'green-glass', 'metal', 'paper', 'plastic', 'shoes', 'trash', 'white-glass']
```

```python
import math

num_images = len(lists)
num_rows = math.ceil(num_images / 4)

fig, axes = plt.subplots(num_rows, 4, figsize=(15, 5*num_rows))

for i, img_name in enumerate(lists):
    print(image_path + img_name)
    image_input = load_img((image_path + img_name), target_size=(img_height, img_width))

    #  NumPy array
    input_img_array = img_to_array(image_input)

    # 
    input_img_array = input_img_array.astype('float32') / 255

    # 
    input_image = np.expand_dims(input_img_array, axis=0)

    #  CNN
    prediction = classifier.predict(input_image)

    # 
    predicted_label = np.argmax(prediction)
    row_index = i // 4
    col_index = i % 4
    axes[row_index, col_index].imshow(image_input)
    axes[row_index, col_index].set_title("Predicted Label: {}".format(categories[predicted_label]))
    axes[row_index, col_index].axis('off')

plt.tight_layout()
plt.show()
```
